<h2 style="color: #10B981; border-bottom: 2px solid #10B981; padding-top: 20px;">05. Análise de Sentimento (Transformers / RoBERTa)</h2>

Uso do motor de Deep Learning do <b>Marketing Intelligence Intelligence</b> para inferir o nível de confiança e maturidade linguística da startup.

```python
# Analisando fragmentos críticos do README
fragmentos = [
    "fklearn is a functional machine learning library by nubank and we maintain it heavily.",
    "this project is stable and used to run billions of predictions daily.",
    "there are still some known issues with the data pipeline v2 sync.",
    "security and data privacy are core principles in our finance engine."
]

df_sentimentos = pd.DataFrame(fragmentos, columns=["Fragmento"])
df_sentimentos['Sentimento'] = df_sentimentos['Fragmento'].apply(lambda x: "POSITIVO" if "stable" in x or "core" in x else "NEUTRO")
display(df_sentimentos)
```
