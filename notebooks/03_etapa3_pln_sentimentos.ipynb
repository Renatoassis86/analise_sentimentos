<h2 style="color: #6366F1; border-bottom: 2px solid #6366F1; padding-top: 40px; font-family: 'Outfit';">01. Configuração do Ecossistema MiNER</h2>

A configuração do ambiente prioriza o modelo <b>pt_core_news_sm</b> para esta demonstração dinâmica, garantindo baixa latência de resposta, embora o sistema de produção suporte <i>large models</i> para auditorias de conformidade legal.

```python
import pandas as pd
import numpy as np
import spacy
import requests
import time
from spacy import displacy
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
plt.style.use('ggplot')
sns.set_theme(style="whitegrid", palette="pastel")

print("📡 Inicializando barramento de dados PLN...")

try:
    nlp = spacy.load("pt_core_news_sm")
    print("✅ Motor spaCy (sm) Ativo.")
except:
    import os
    os.system("python -m spacy download pt_core_news_sm")
    nlp = spacy.load("pt_core_news_sm")

print("🚀 MiNER v14.4 Engine: Ready for Deep Audit.")
```

<h2 style="color: #6366F1; border-bottom: 2px solid #6366F1; padding-top: 40px; font-family: 'Outfit';">02. Ingestão Dinâmica de Dados (Demonstração N=5)</h2>

Nesta fase, simulamos a conexão com o <b>Data Lake</b> do projeto. Buscamos documentações reais de repositórios oficiais para validar a transparência algorítmica das Fintechs.

```python
fintechs_alvo = [
    {"nome": "Nubank", "slug": "nubank/fklearn", "url": "https://raw.githubusercontent.com/nubank/fklearn/master/README.md"},
    {"nome": "Stone", "slug": "stone-payments/pede-direto", "url": "https://raw.githubusercontent.com/stone-payments/pede-direto/master/README.md"},
    {"nome": "Stark Bank", "slug": "starkbank/starkbank-python", "url": "https://raw.githubusercontent.com/starkbank/starkbank-python/master/README.md"},
    {"nome": "Ebanx", "slug": "ebanx/integration-sample", "url": "https://raw.githubusercontent.com/ebanx/integration-sample/master/README.md"},
    {"nome": "Banco Inter", "slug": "banco-inter/api-java", "url": "https://raw.githubusercontent.com/banco-inter/api-java/master/README.md"}
]

def extrair_readme(fintech):
    try:
        r = requests.get(fintech['url'], timeout=10)
        if r.status_code == 200:
            return r.text[:1500] # Pegamos os primeiros 1500 chars para a demonstração
        return "Documentação não acessível publicamente via API."
    except:
        return "Erro de conexão com o repositório."

print("📥 Iniciando ingestão via Auditoria API...")
df_injestao = pd.DataFrame(fintechs_alvo)
df_injestao['conteudo_cru'] = df_injestao.apply(extrair_readme, axis=1)

print(f"✅ Ingestão finalizada. {len(df_injestao)} fontes processadas.")
display(df_injestao[['nome', 'slug', 'conteudo_cru']].head())
```

<h2 style="color: #6366F1; border-bottom: 2px solid #6366F1; padding-top: 40px; font-family: 'Outfit';">03. Metodologia: Auditoria Gramatical (POS Tagging)</h2>

### Teoria Acadêmica
A **Auditoria Gramatical** via Part-of-Speech (POS) Tagging é fundamentada na hipótese de que a maturidade de um ecossistema de dados se reflete na sua linguagem técnica. 
- **Verbos (VERB):** Indicam capacidade de execução e funções algorítmicas.
- **Substantivos (NOUN):** Representam os ativos de dados e entidades do sistema.
- **Adjetivos (ADJ):** Se excessivos, denotam subjetividade comercial ("Marketing Fluff").

Analizamos a razão entre Verbos e Adjetivos para medir a **Objetividade Sistêmica**.

```python
def processar_gramatica(texto):
    doc = nlp(texto)
    pos_counts = {"VERB": 0, "NOUN": 0, "ADJ": 0, "ADV": 0}
    for token in doc:
        if token.pos_ in pos_counts:
            pos_counts[token.pos_] += 1
    return pos_counts

df_gramatica = df_injestao.copy()
df_gramatica['pos_stats'] = df_gramatica['conteudo_cru'].apply(processar_gramatica)

# Expandindo o dicionário em colunas
df_pos_expanded = df_gramatica['pos_stats'].apply(pd.Series)
df_final_gramatica = pd.concat([df_gramatica['nome'], df_pos_expanded], axis=1)

print("📊 Estrutura Gramatical Consolidada:")
display(df_final_gramatica)

# Visualização da Maturidade de Documentação
df_final_gramatica.set_index('nome').plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')
plt.title("Maturidade Gramatical: Análise de Verbos vs Substantivos")
plt.ylabel("Frequência de Tags")
plt.show()
```

<h2 style="color: #6366F1; border-bottom: 2px solid #6366F1; padding-top: 40px; font-family: 'Outfit';">04. Reconhecimento de Entidades (NER) e Governança</h2>

### Teoria de Governança Digital
O mapeamento de **Named Entities (NER)** nos permite identificar automaticamente quais frameworks de segurança (ISO, PCI-DSS) ou organizações externas (BCB, CVM) são citados na documentação. Isso valida o pilar de <b>Compliance e Atividade</b>.

Visualizamos abaixo as conexões extraídas de um manifesto de segurança.

```python
exemplo_governança = """
A Stark Bank segue as diretrizes do Banco Central do Brasil e utiliza protocolos da Amazon Web Services 
em sua sede em São Paulo para garantir a segurança dos dados via criptografia AES-256.
"""
doc_ner = nlp(exemplo_governança)

print("📍 Mapeamento de Entidades Críticas:")
displacy.render(doc_ner, style="ent", jupyter=True)

# Extração tabular para o Score
entities_data = [{"Texto": ent.text, "Rótulo": ent.label_} for ent in doc_ner.ents]
display(pd.DataFrame(entities_data))
```

<h2 style="color: #6366F1; border-bottom: 2px solid #6366F1; padding-top: 40px; font-family: 'Outfit';">05. Auditoria Profunda via Transformers (Inferência Contextual)</h2>

### O Salto para o Contexto
Diferente da Etapa 1 (Regex), os **Transformers** não buscam apenas palavras, mas a <b>intenção semântica</b>. 
- <b>Falso Positivo:</b> "This is not a bad security system" (Regex focaria em "bad").
- <b>Verdadeira Intenção:</b> O Transformer entende a negação e classifica como Positivo.

Aqui simulamos a latência e o resultado do motor <b>RoBERTa-base-pt</b>.

```python
def motor_transformer_simulado(texto):
    # Simulação de pipeline(task='sentiment-analysis', model='roberta-pt-fintech')
    time.sleep(0.5) # Simula a carga computacional de um Transformer
    palavras = len(texto.split())
    if "error" in texto.lower() or "risk" in texto.lower():
        return {"label": "RISCO", "score": np.random.uniform(0.85, 0.99)}
    return {"label": "SAUDÁVEL", "score": np.random.uniform(0.90, 0.99)}

print("🤖 Iniciando Inferência em Deep Learning...")
df_injestao['inferencia_transformer'] = df_injestao['conteudo_cru'].apply(motor_transformer_simulado)
df_transformer = pd.json_normalize(df_injestao['inferencia_transformer'])
df_final_audit = pd.concat([df_injestao['nome'], df_transformer], axis=1)

display(df_final_audit)
```

<h2 style="color: #6366F1; border-bottom: 2px solid #6366F1; padding-top: 40px; font-family: 'Outfit';">06. Consolidação: Cálculo do Score SHA 4D (Algoritmo v14)</h2>

Combinamos todos os sinais capturados (Gramática, NER, Transformers) para gerar o **Health Score** final que alimenta o Dashboard ArKOS.

```python
def calcular_score_sha_v14(row):
    # Fatores Ponderados
    audit_dl = 100 if row['label'] == 'SAUDÁVEL' else 40
    confianca = row['score'] * 100
    
    # Cálculo Proprietário
    score_final = (audit_dl * 0.6) + (confianca * 0.4)
    return round(score_final, 2)

df_ranking = df_final_audit.copy()
df_ranking['SHA_4D_Score'] = df_ranking.apply(calcular_score_sha_v14, axis=1)
df_ranking = df_ranking.sort_values(by='SHA_4D_Score', ascending=False)

print("🏆 RANKING FINAL DE AUDITORIA — ETAPA 3:")
display(df_ranking)

# Gráfico de Radar ou Barras Técnico
plt.figure(figsize=(10, 5))
colors = ['#10B981' if s > 85 else '#F59E0B' for s in df_ranking['SHA_4D_Score']]
plt.bar(df_ranking['nome'], df_ranking['SHA_4D_Score'], color=colors)
plt.axhline(80, color='red', linestyle='--', label='Limiar de Alerta')
plt.title("Health Score Final: Auditoria de Marketing Intelligence")
plt.ylabel("Score SHA (%)")
plt.legend()
plt.show()
```
